```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#points to be plotted
x=[2,3,5]
y=[1,2,4]
#giving title to the graph
plt.title("graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#resize graph
plt.figure(figsize=(10,5),dpi=130)

#plotting the graph
plt.plot(x,y,label="y=2x",color="red", linewidth=2,linestyle="--",marker="",markeredgecolor="blue")

#to show the label we put on which is y=2x
plt.legend()#to show the label we put on which is y=2x
plt.grid(True)

#scaling or sizing of graph by changing its axis
plt.axis([0,6,0,6])
plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,1,2,3,4,5,6])

# plt.savefig("mygraph.png",dpi=300)
plt.show()
```

###barchart


```python
labels=[1,2,3,4,5]
values=[10,20,30,40,50]

bars=plt.bar(labels,values,color="red")
plt.title("barchart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
bars[0].set_color("green")
bars[1].set_hatch('/')
bars[2].set_color("blue")
plt.grid(True)
plt.figure(figsize=(6,4))

plt.show()

```

###histogram


```python
x=[1,2,34,5,6,7,8,9,10]
plt.hist(x,bins=5,color="red")
plt.title("histogram")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line chart
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')

# Add annotations
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

# Add title and labels
plt.title('Line Chart with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```
